```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
url = 'https://raw.githubusercontent.com/Srieswari/CuisineClassification/main/indian_food.csv'
df = pd.read_csv(url)
print(df.shape)
df.head(3)
```

```python
def find_recipes(cuisine, ingredients, n):
    cuisine_filtered = df[df['Cuisine'].str.contains(cuisine, case=False, na=False)]
    ingredient_filtered = cuisine_filtered[cuisine_filtered['TranslatedInstructions'].apply(lambda x: all(ingredient in x for ingredient in ingredients))].head(n)

    return ingredient_filtered[['TranslatedInstructions', 'URL']]
```

```python
def format_response(TranslatedInstructions):
    if TranslatedInstructions.empty:
        return "Sorry, no recipes found matching your criteria."
    response = "Here are some recipes you might like:\n\n"
    for idx, row in TranslatedInstructions.iterrows():
        response += f"{row['TranslatedInstructions']}\nLink: {row['URL']}\n\n"
    return response
```

```python
def chat():
    print("Welcome to the Recipe Chatbot!")
    while True:
        cuisine = input("Enter a cuisine type (or 'exit' to quit): ").strip().lower()
        if cuisine == 'exit':
            break
        ingredients = input("Enter ingredients separated by commas: ").strip().lower().split(',')
        ingredients = [ingredient.strip() for ingredient in ingredients]
        n = int(input("Enter the number of recipes you want to see: "))
        TranslatedInstructions = find_recipes(cuisine, ingredients, n)
        response = format_response(TranslatedInstructions)
        if len(TranslatedInstructions)<n:
          print (f"Sorry only {len(TranslatedInstructions)} found")
        print(response)
```

```python
chat()
```
